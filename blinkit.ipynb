# ***`Handling Missing values`***

```python
data.info()
```

```python
Item_Weight_mean=data.groupby('Item_Type')['Item_Weight'].mean()
display(Item_Weight_mean)
```

```python
data['Item_Weight']=data.apply(lambda x:Item_Weight_mean[x['Item_Type']] if pd.isnull(x['Item_Weight']) else x['Item_Weight'],axis=1)
data.info()
```



---



```python
null_rows = data[data['Outlet_Size'].isnull()]
display(null_rows)
```



---



```python
# Calculate the mode of 'Outlet_Size' for each group
outlet_size_mode = data.groupby(['Outlet_Location_Type', 'Outlet_Type'])['Outlet_Size'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)

# Reset the index of the result to have a single index
# Instead of creating a DataFrame, create a Series with 'Outlet_Location_Type' and 'Outlet_Type' as index
outlet_size_mode = outlet_size_mode.reset_index().set_index(['Outlet_Location_Type', 'Outlet_Type'])['Outlet_Size']
display(outlet_size_mode)

# Update the 'Outlet_Size' column with the mode values, handling NaN and KeyError
# Use the Series 'outlet_size_mode' to fill NaN values in 'Outlet_Size'
# If KeyError occurs, keep the original 'Outlet_Size' value
data['Outlet_Size'] = data.apply( lambda x: outlet_size_mode.get((x['Outlet_Location_Type'], x['Outlet_Type']))
    if pd.isnull(x['Outlet_Size']) and (x['Outlet_Location_Type'], x['Outlet_Type']) in outlet_size_mode.index
    else x['Outlet_Size'],
    axis=1
)
```

```python
# Fill any remaining null values in 'Outlet_Size' with 'Small', assuming it's a frequent value
data['Outlet_Size']=data['Outlet_Size'].fillna('Small')
```

# ***`Standardize Categorical Data`***

```python
data['Item_Fat_Content'].value_counts()
```

```python
data['Item_Fat_Content']=data['Item_Fat_Content'].str.replace('LF','Low Fat')
data['Item_Fat_Content']=data['Item_Fat_Content'].str.replace('reg','Regular')
data['Item_Fat_Content']=data['Item_Fat_Content'].str.replace('low fat','Low Fat')
```

# ***`Remove Outliers`***

```python
# Calculate the mean visibility of items with visibility greater than 0
mean_visibility = data[data['Item_Visibility'] > 0]['Item_Visibility'].mean()
display(mean_visibility)

# Replace 0 values in 'Item_Visibility' with the calculated mean
data['Item_Visibility'] = data['Item_Visibility'].replace(0, mean_visibility)
```

```python
# Count rows where Item_Visibility is 0
count = data[data['Item_Visibility'] == 0]['Item_Visibility'].count()
print(f"Number of rows with Item_Visibility=0: {count}")
```

# ***`Create Derived Field`***

# ***`Column1 : Sales_Per_Item`***

```python
data['Sales_Per_Item']=data['Item_Outlet_Sales']/data['Item_MRP']
display(data)
```



---



# ***`Column2: Outlet_Age`***

```python
# Get the current year
Current_Year=pd.to_datetime('today').year

# Subtract the Outlet Establishment Year from the current year to get the Outlet Age
Outlet_Age= Current_Year - data['Outlet_Establishment_Year']

# Add the Outlet Age as a new column to the dataframe
data['Outlet_Age']=Outlet_Age
display(data)
```

# ***`Normalised Numerical Data`***

```python
# Define the function as categorize_price
def categorize_price(price):
    if price <= 100:
        return "Low"
    elif price <= 200:
        return "Medium"
    else:
        return "High"

# Apply the categorize_price function to the 'Item_MRP' column and create a new column 'Price_Tier'
data['Price_Tier'] = data['Item_MRP'].apply(categorize_price)

# View the Results
display(data[['Item_MRP', 'Price_Tier']].head())
```

# ***`Adding Descriptive Labels`***

```python
data['Outlet_Type'].value_counts()
```

```python
# Define the function named as Outlet_Type_Labels
def Outlet_Type_Labels(outlet_type):
    if outlet_type =='Supermarket Type1':
      return "Supermarket Type 1"
    elif outlet_type =='Supermarket Type2':
      return "Supermarket Type 2"
    elif outlet_type =='Supermarket Type3':
      return "Supermarket Type 3"
    else:
      return "Local Grocery Store"

# Update the function on the "Outlet_Type" column
data['Outlet_Type']=data['Outlet_Type'].apply(Outlet_Type_Labels)

# View the Results
display(data['Outlet_Type'].value_counts())
```

```python
data['Outlet_Location_Type'].value_counts()
```

```python
# Define the function named as Outlet_Location_Labels
def Outlet_Location_Labels(outlet_location):
    if outlet_location =='Tier 1':
      return "Urban Area"
    elif outlet_location =='Tier 2':
      return "Sub-urban Area"
    elif outlet_location =='Tier 3':
      return "Rural Area"
    else:
      return "Unknown"

# Update the function on the "Outlet_Location_Type" column
data['Outlet_Location_Type']=data['Outlet_Location_Type'].apply(Outlet_Location_Labels)

# View the Results
display(data['Outlet_Location_Type'].value_counts())
```

# ***`Verify Unique Identifiers`***

```python
# Investigate count of Item_Identifier Reused
item_counts = data.groupby('Item_Identifier')['Item_Identifier'].count().reset_index(name='count')
duplicates = item_counts[item_counts['count'] > 1].sort_values(by='count', ascending=False)
result = duplicates[['Item_Identifier', 'count']]
print(result)
```

```python
# Checking for duplicates
duplicates=data[data.duplicated(subset='Item_Identifier',keep='first')]
print(f"Number of duplicate item_identifiers: {len(duplicates)}")
display(duplicates)
```

```python
# Cleaning the data
Data=data.groupby('Item_Identifier').agg({'Item_MRP':'max','Item_Weight':'max'}).reset_index()
display(Data)
```

```python
# Verify Uniqueness
assert Data['Item_Identifier'].is_unique, "Item identifiers are still not unique!"
print("All item_identifiers are unique.")
```

# ***`Handle Data Type Consistency`***

```python
data['Outlet_Establishment_Year']=pd.Categorical(data['Outlet_Establishment_Year'])
```

```python
# Define the function as outlet_age
def outlet_age(row):
  current_year = pd.to_datetime('today').year
  establishment_year = row['Outlet_Establishment_Year']

  # Convert establishment_year to int before subtraction
  age = current_year - int(establishment_year)
  if age <= 25:
    return 'Old'
  elif  25 < age <=35:
    return 'Moderate'
  else:
    return 'New'

# Creating new column as 'Outlet_Categorical_Age'
data['Outlet_Categorical_Age'] = data.apply(outlet_age, axis=1)

# Display the Results
display(data)
```

```python
data.groupby('Outlet_Age')['Outlet_Categorical_Age'].unique()
```
